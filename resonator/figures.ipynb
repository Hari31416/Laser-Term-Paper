## Using Parametric Equations

The parmetric equation of circle is:
$$
x - x_0 = r \cos \theta \\
y - y_0 = r \sin \theta
$$

### Concave Mirror

For $x_0<0$

```python
theta1 = 130
theta2 = 230
thetas = np.linspace(theta1, theta2, 100)

fig, ax = plt.subplots()
a, b = 1,1 
x0 = -1
y0 = 0

x = x0 + a
# y = y0
xs = a*np.cos(np.deg2rad(thetas)) + x
ys = b*np.sin(np.deg2rad(thetas)) + y0
ax.plot(xs, ys)
ax.axvline(x = x0, color = 'k')

angles = np.arctan2(ys, xs)
angles = np.abs(angles)
mask = ys <= y0
angles[mask] = -angles[mask]
#Drawing small lines to the mirror
for i in range(0, len(xs), 5):
    ax.plot([xs[i], xs[i] + 0.1*np.cos((angles[i]))], [ys[i], ys[i] + 0.1*np.sin((angles[i]))], color = 'k')

plt.xlim(-2, 2)
plt.ylim(-2, 2)
```

for $x_0>0$

```python
fig, ax = plt.subplots()

theta1 = 130
theta2 = 230
thetas = np.linspace(theta1, theta2, 100)

a, b = 1,1
x0 = 1
y0 = 0

x = x0 - a
y = y0
xs = -a*np.cos(np.deg2rad(thetas)) + x
ys = -b*np.sin(np.deg2rad(thetas)) + y

ax.plot(xs, ys)
ax.axvline(x = x0, color = 'k')

angles = np.arctan2(ys, xs)
angles = np.abs(angles)
mask = ys < 0
angles[mask] = -angles[mask]
#Drawing small lines to the mirror
for i in range(0, len(xs), 5):
    ax.plot([xs[i], xs[i] + 0.1*np.cos((angles[i]))], [ys[i], ys[i] + 0.1*np.sin((angles[i]))], color = 'k')



ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2);
```

### Final Functions

```python
def concave_mirror(ax, x0, y0, a, b, hline = False):
    theta1 = 130
    theta2 = 230
    thetas = np.linspace(theta1, theta2, 100)

    if x0<0:
        x = x0 + a
        xs = a*np.cos(np.deg2rad(thetas)) + x
        ys = b*np.sin(np.deg2rad(thetas)) + y0

    else:
        x = x0 - a
        xs = -a*np.cos(np.deg2rad(thetas)) + x
        ys = -b*np.sin(np.deg2rad(thetas)) + y0

    ax.plot(xs, ys)
    if hline:
        ax.axvline(x = x0, color = 'k')

    angles = np.arctan2(ys, xs)
    angles = np.abs(angles)
    if a<0:
        mask = ys >= y0
        
    else:
        mask = ys <= y0
    angles[mask] = -angles[mask]
    

    #Drawing small lines to the mirror
    for i in range(0, len(xs), 5):
        p1 = [xs[i], xs[i] + 0.1*np.cos((angles[i]))]
        p2 = [ys[i], ys[i] + 0.1*np.sin((angles[i]))]
        ax.plot(p1, p2, color = 'k')
```

```python
fig, ax = plt.subplots()
concave_mirror(ax, -1.5, 0, 1, 1, True)
# concave_mirror(ax, 1.5, 0, -1, -1, True)

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2);
```

```python
def convex_mirror(ax, x0, y0, a, b, hline = False):
    theta1 = 130
    theta2 = 230
    thetas = np.linspace(theta1, theta2, 100)

    if x0<0:
        x = x0 - a
        xs = -a*np.cos(np.deg2rad(thetas)) + x
        ys = -b*np.sin(np.deg2rad(thetas)) + y0

    else:
        x = x0 + a
        xs = a*np.cos(np.deg2rad(thetas)) + x
        ys = b*np.sin(np.deg2rad(thetas)) + y0

    ax.plot(xs, ys)
    if hline:
        ax.axvline(x = x0, color = 'k')

    angles = np.arctan2(ys, xs)
    angles = np.abs(angles)
    mask = ys >= y0
    angles[mask] = -angles[mask]

    #Drawing small lines to the mirror
    for i in range(0, len(xs), 5):
        p1 = [xs[i], xs[i] + 0.1*np.cos((angles[i]))]
        p2 = [ys[i], ys[i] + 0.1*np.sin((angles[i]))]
        ax.plot(p1, p2, color = 'k')
```

```python
fig, ax = plt.subplots()
convex_mirror(ax, -1.5, 0, 1, 2, True)
convex_mirror(ax, 1.5, 0, 1, 1, True)
```

```python
def get_x(y, y0, r, x0):
    if x0<0:
        sintheta = (y - y0)/r
        x = x0 + r*np.cos(np.arcsin(sintheta))
        return -(x + r)
    else:
        sintheta = (y - y0)/r
        x = x0 - r*np.cos(np.arcsin(sintheta))
        return x + r
```

# Drawing Rays

Here, we'll draw the rays which will be used in the ray tracing.

```python
length = 1.5
theta = 30
start_point = [0, 0]
end_point = [length*np.cos(np.deg2rad(theta)), length*np.sin(np.deg2rad(theta))]
x_points = np.linspace(start_point[0], end_point[0], 100)
y_points = np.linspace(start_point[1], end_point[1], 100)

fig, ax = plt.subplots()
ax.plot(x_points, y_points, color = 'k')
#Now the arrow
ax.arrow(end_point[0], end_point[1], 0.1*np.cos(np.deg2rad(theta)), 0.1*np.sin(np.deg2rad(theta)), head_width=0.1, head_length=0.2, fc='k', ec='k')

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2);
```

```python
def plot_arrow(ax, start_point, length, theta, return_end=False):
    start_x = start_point[0]
    start_y = start_point[1]
    end_x = start_x+length*np.cos(np.deg2rad(theta))
    end_y = start_y+length*np.sin(np.deg2rad(theta))
    
    x_points = np.linspace(start_x, end_x, 100)
    y_points = np.linspace(start_y, end_y, 100)
    ax.plot(x_points, y_points, color = 'r')

    dx = 0.01*np.cos(np.deg2rad(theta))
    dy = 0.01*np.sin(np.deg2rad(theta))
    ax.arrow(end_x, end_y, dx, dy, head_width=0.1, head_length=0.1, fc='r', ec='r', length_includes_head=True)
    if return_end:
        return [end_x, end_y]
```

```python
fig, ax = plt.subplots()
p = plot_arrow(ax, [1, 1], 1, 30, True)
end_arrow_2 = plot_arrow(ax, p, 1.5, -180, True)
plot_arrow(ax, end_arrow_2, 1, -90, False)

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2);
```
