```python
%%manim -v WARNING -s -qh --disable_caching RTM_Box

class RTM_Box(Scene):
    def construct(self):
        self.create_frame()
        self.create_arrows()
        self.draw_ys()
    def create_frame(self):
        rectangle = Rectangle(color=RECT_COLOR, height=5, width=3)
        zero_line = DashedLine(4*LEFT, 4*RIGHT, color=DASH_COLOR,)
        self.zero_line = zero_line
        l1 = DashedLine(start = 3*LEFT, end = 1*RIGHT, color=DASH_COLOR)
        l1.move_to(1.5*UP + 1.5*LEFT)
        l2 = DashedLine(start = 3*LEFT, end = 1*RIGHT, color=DASH_COLOR)
        l2.move_to(1*UP + 1.5*RIGHT)
        self.l1 = l1
        self.l2 = l2
        self.add(zero_line)
        self.add(rectangle)
        self.add(l1)
        self.add(l2)

    def create_arrows(self):
        #Create a arrow passing through the center of the line l1
        arrow = Arrow(start = 5*LEFT + 1*DOWN, end = self.l1.get_center(), color=ARROW_COLOR, buff=0)
        self.add(arrow)
        #Draw an angle between the arrow and the line l1 the angle will be in third quadrant
        angle = Angle(arrow, self.l1, radius=1, quadrant=(-1,-1), color=TEXT_COLOR, other_angle=True)
        self.add(angle)
        #Now annotate this angle by theta_i
        theta_0 = MathTex(r"\theta_i", color=TEXT_COLOR)
        theta_0.next_to(angle, buff=-0.6)
        self.add(theta_0)

        #Create a arrow passing through the center of the line l2 and annotate it by theta_r
        arrow = Arrow(end = 5*RIGHT + 1*DOWN, start = self.l2.get_center(), color=ARROW_COLOR, buff=0)
        self.add(arrow)
        angle = Angle(arrow, self.l2, radius=1, quadrant=(1,1), color=TEXT_COLOR, other_angle=False)
        self.add(angle)
        theta_r = MathTex(r"\theta_r", color=TEXT_COLOR)
        theta_r.next_to(angle, buff=0.1)
        self.add(theta_r)

    def draw_ys(self):
        #draw an dashed vertical arrow passing through the center of the line l1
        y1 = DoubleArrow(start = ORIGIN - 1.5*RIGHT, end = self.l1.get_center(), color=TEXT_COLOR, buff=0)
        y1 = DashedVMobject(y1)
        y1.shift(0.2*RIGHT)
        self.add(y1)
        #annotate the arrow with y_0
        y_i = MathTex(r"y_i", color=TEXT_COLOR)
        y_i.next_to(y1, buff=0.1)
        self.add(y_i)

        y2 = DoubleArrow(start = ORIGIN + 1.5*RIGHT, end = self.l2.get_center(), color=TEXT_COLOR, buff=0)
        y2 = DashedVMobject(y2)
        y2.shift(0.2*LEFT)
        self.add(y2)
        y_r = MathTex(r"y_r", color=TEXT_COLOR)
        y_r.next_to(y2, buff=-0.7)
        self.add(y_r)
```

```python
%%manim -v WARNING -s -qh --disable_caching Sign_Convention_1

class Sign_Convention_1(Scene):
    def construct(self):
        self.create_frame()
        self.create_arrows()
        self.draw_ys()
        all_mobjects = self.mobjects
        all_mob = VGroup(*all_mobjects)
        all_mob.scale(1.3)
    def create_frame(self):
        zero_line = DashedLine(5*LEFT, 5*RIGHT, color=DASH_COLOR,)
        self.zero_line = zero_line
        zero = Text("Zero Line", color=TEXT_COLOR).scale(0.5)
        zero.next_to(zero_line).shift(2*LEFT + 0.3*DOWN)
        self.add(zero)
        l1 = DashedLine(start = 3*LEFT, end = 1*RIGHT, color=DASH_COLOR)
        l1.move_to(1.5*UP + 1.5*LEFT)
        l2 = DashedLine(start = 3*LEFT, end = 1*RIGHT, color=DASH_COLOR)
        l2.move_to(-1*UP + 1.5*RIGHT)
        self.l1 = l1
        self.l2 = l2
        self.add(zero_line)
        self.add(l1)
        self.add(l2)

    def create_arrows(self):
        #Create a arrow passing through the center of the line l1
        arrow = Arrow(start = 5*LEFT + 1*DOWN, end = self.l1.get_center(), color=ARROW_COLOR, buff=0)
        self.add(arrow)

        #Create a arrow passing through the center of the line l2 and annotate it by theta_r
        arrow = Arrow(end = 5*RIGHT + 1*UP, start = self.l2.get_center(), color=ARROW_COLOR, buff=0)
        self.add(arrow)

    def draw_ys(self):
        #draw an dashed vertical arrow passing through the center of the line l1
        y1 = DoubleArrow(start = ORIGIN - 1.5*RIGHT, end = self.l1.get_center(), color=TEXT_COLOR, buff=0)
        y1 = DashedVMobject(y1)
        y1.shift(0.2*RIGHT)
        self.add(y1)
        v1 = Line(start = ORIGIN - 1.5*RIGHT + 1*DOWN, end = self.l1.get_center() + 1*UP, color=TEXT_COLOR)
        self.add(v1)
        #annotate the arrow with y_0
        y_i = MathTex(r"y_i", color=TEXT_COLOR)
        positive = Text("Positive", color = TEXT_COLOR).scale(0.5)
        y_i.next_to(y1, buff=0.1)
        positive.next_to(y_i)
        self.add(y_i)
        self.add(positive)

        y2 = DoubleArrow(start = ORIGIN + 1.5*RIGHT, end = self.l2.get_center(), color=TEXT_COLOR, buff=0)
        y2 = DashedVMobject(y2)
        y2.shift(0.2*LEFT)
        self.add(y2)
        v2 = Line(start = ORIGIN + 1.5*RIGHT + 1*UP, end = self.l2.get_center() + 1*DOWN, color=TEXT_COLOR)
        self.add(v2)
        y_r = MathTex(r"y_r", color=TEXT_COLOR)
        y_r.next_to(y2, buff=-0.7)
        negative = Text("Negative", color = TEXT_COLOR).scale(0.5)
        negative.next_to(y_r).shift(2.5*LEFT)
        self.add(y_r)
        self.add(negative)
```

```python
%%manim -v WARNING -s -qh --disable_caching SMR
MIRROR_COLOR = RED
class SMR(Scene):
    def construct(self):
        self.create_mirrors()
        self.create_rays()
        all_mobjects = self.mobjects
        all_mob = VGroup(*all_mobjects)
        all_mob.scale(1.3)

    def create_mirrors(self):
        #Create a caoncave mirror to the left
        left_mirror = self.concave_mirror(-4, 0, 3, "M_1")
        right_mirror = self.concave_mirror(4, 0, -3, "M_2")
        self.left_mirror = left_mirror
        self.right_mirror = right_mirror
        self.add(right_mirror)
        self.add(left_mirror)
        zero_line = DashedLine(start = 5*LEFT, end = 5*RIGHT, color = DASH_COLOR)
        #Create a double vector for the length of the mirror
        arrow = DoubleArrow(start = 4.5*LEFT, end = 4.5*RIGHT, color=BLACK, buff=0).shift(0.2*DOWN)
        length = MathTex(r"L", color=TEXT_COLOR)
        length.next_to(arrow.get_center(), buff=0.1).scale(0.7).shift(0.3*UP)
        self.add(length)
        self.add(arrow)
        self.add(zero_line)

    def concave_mirror(self, x, y, R, name):
        #Creates a concave mirror of radius R at position x, y
        mirror = Arc(radius=R, start_angle=130*np.pi/180, angle=100*np.pi/180, color=MIRROR_COLOR)
        mirror.move_to(x*RIGHT + y*UP)
        #add the name
        name = MathTex(name, color=TEXT_COLOR)
        name.next_to(mirror.get_end(), buff=0).scale(0.7)

        mirror = VGroup(mirror, name)
        return mirror

    def create_rays(self):
        #Create the first ray, moving from the left mirror to the right mirror
        ray_strt_pos = 4.4*LEFT + 1.5*UP
        ray_end_pos = 4.8*RIGHT + 0.2*UP
        ray = Arrow(start = ray_strt_pos, end = ray_end_pos, color = ARROW_COLOR)
        self.add(ray)
        annotate1 = Text("(1) Translation from left mirror to right mirror", color = TEXT_COLOR).scale(0.3)
        annotate1.next_to(ray.get_center() + 0.3*UP + 1*LEFT, buff = 0.5)
        self.add(annotate1)

        annotate2 = Text("(2) Reflection from right mirror", color = TEXT_COLOR).scale(0.3).rotate(PI/2)
        annotate2.next_to(ray.get_end(), buff = 0.2)
        self.add(annotate2)

        dashed_line = DashedLine(start = ray.get_end(), end = ray.get_end() + 3*RIGHT, color = DASH_COLOR).shift(2*LEFT)
        self.add(dashed_line)

        ray_strt_pos = ray.get_end()
        ray_end_pos = 4.6*LEFT - 1.1*UP
        ray = Arrow(start = ray_strt_pos, end = ray_end_pos, color = ARROW_COLOR)
        self.add(ray)
        annotate2 = Text("(3) Translation from right mirror to left mirror", color = TEXT_COLOR).scale(0.3)
        annotate2.next_to(ray.get_center() + 0.3*DOWN + 2*LEFT, buff = 0.5)
        self.add(annotate2)
        annotate4 = Text("(4) Reflection from left mirror", color = TEXT_COLOR).scale(0.3).rotate(PI/2)
        annotate4.next_to(ray.get_end(), buff = -0.4)
        self.add(annotate4)
        dashed_line = DashedLine(start = ray.get_end(), end = ray.get_end() + 3*RIGHT, color = DASH_COLOR).shift(RIGHT*LEFT)
        self.add(dashed_line)
```

```python
%%manim -v WARNING -s -qk --disable_caching Translation
MIRROR_COLOR = RED
class Translation(Scene):
    def construct(self):
        self.create_plane_mirrors()
        self.create_rays()
    
    def create_plane_mirrors(self):
        left_mirror = Line(start = 3*DOWN, end = 3*UP, color = MIRROR_COLOR,).shift(4*LEFT)
        right_mirror = Line(start = 3*DOWN, end = 3*UP, color = MIRROR_COLOR).shift(4*RIGHT)
        zero_line = DashedLine(start = 5*LEFT, end = 5*RIGHT, color = DASH_COLOR)
        self.add(zero_line)
        self.add(left_mirror)
        self.add(right_mirror)

    def create_rays(self):
        ray_start = 4.2*LEFT + 0.5*UP
        ray_end = 4.3*RIGHT + 2.5*UP
        ray = Arrow(start = ray_start, end = ray_end, color = ARROW_COLOR)
        hline_1 = DashedLine(start = ORIGIN, end = 3*RIGHT, color = DASH_COLOR)
        hline_1.next_to(ray.get_start(), buff = 0).shift(1.5*LEFT)
        #Create theta between the ray and the horizontal line
        angle = Angle(hline_1, ray, radius = 1, color = TEXT_COLOR)
        self.add(angle)
        theta = MathTex(r"\theta_i", color = TEXT_COLOR).scale(0.7)
        theta.next_to(angle, buff = 0.2)
        #Create a brace to measure the y_i
        y_i_text = MathTex(r"y_i", color = TEXT_COLOR).scale(0.8)
        y_i_text.next_to(4*LEFT + 0.3*UP, buff = -0.6)

        f_arrow_start = 4.2*LEFT + 0.3*DOWN
        f_arrow_end = 4.2*LEFT + 0.8*UP
        f_arrow = DoubleArrow(start = f_arrow_start, end = f_arrow_end, color = BLACK)
        self.add(theta)
        self.add(ray)
        self.add(hline_1)
        self.add(y_i_text)
        self.add(f_arrow)

        hline_2 = DashedLine(start = ORIGIN, end = 3*RIGHT, color = DASH_COLOR)
        hline_2.next_to(ray.get_end(), buff = 0).shift(1.5*LEFT)
        #Create theta between the ray and the horizontal line
        angle = Angle(hline_2, ray, radius = 1, color = TEXT_COLOR, other_angle = False, quadrant = (-1, -1))
        self.add(angle)
        theta = MathTex(r"\theta_f", color = TEXT_COLOR).scale(0.7)
        theta.next_to(angle, buff = -0.4)
        #Create a brace to measure the y_i
        y_r_text = MathTex(r"y_f", color = TEXT_COLOR).scale(0.8)
        y_r_text.next_to(4*RIGHT + 1.3*UP, buff = 0.4)

        f_arrow_start = 4.2*RIGHT + 0.3*DOWN
        f_arrow_end = 4.2*RIGHT + 2.8*UP
        f_arrow = DoubleArrow(start = f_arrow_start, end = f_arrow_end, color = BLACK)
        self.add(f_arrow)
        self.add(theta)
        self.add(hline_2)
        self.add(y_r_text)
```

```python
%%manim -v WARNING -s -qk --disable_caching Reflection
MIRROR_COLOR = RED
class Reflection(Scene):
    def construct(self):
        self.create_mirror()
        self.create_lines()
        all_mobjects = self.mobjects
        all_mob = VGroup(*all_mobjects)
        all_mob.scale(1.25)


    def create_mirror(self):
        self.mirror = self.concave_mirror(4, 0, -4, r"\mathcal{M}")
        self.add(self.mirror)
        self.zero_line = DashedLine(start = 5*LEFT, end = 5*RIGHT, color = DASH_COLOR)
        self.add(self.zero_line)

    def concave_mirror(self, x, y, R, name):
        #Creates a concave mirror of radius R at position x, y
        mirror = Arc(radius=R, start_angle=130*np.pi/180, angle=100*np.pi/180, color=MIRROR_COLOR)
        mirror.move_to(x*RIGHT + y*UP)
        #add the name
        name = MathTex(name, color=TEXT_COLOR)
        name.next_to(mirror.get_end(), buff=0).scale(0.7).shift(0.2*DOWN)

        mirror = VGroup(mirror, name)
        return mirror

    def create_lines(self):
        center = Dot(color = MIRROR_COLOR)
        center_annotation = Text("C", color = TEXT_COLOR).scale(0.5).shift(0.3*DOWN)
        self.add(center)
        self.add(center_annotation)

        ray_1_start = 2*LEFT + 0.1*DOWN
        ray_1_end = 4.2*RIGHT + 2.5*UP
        ray_1 = Arrow(start = ray_1_start, end = ray_1_end, color = ARROW_COLOR)
        self.add(ray_1)
        angle_i = Angle(self.zero_line, ray_1, radius = 1, color = TEXT_COLOR)
        self.add(angle_i)
        angle_i_annotation = MathTex(r"\theta_i", color = TEXT_COLOR).scale(0.7)
        angle_i_annotation.next_to(angle_i, buff = 0.2)
        self.add(angle_i_annotation)


        dashed_line_1 = DashedLine(start = ray_1.get_end(), end = ORIGIN, color = DASH_COLOR)
        self.add(dashed_line_1)
        angle_c = Angle(self.zero_line, dashed_line_1, radius = 0.5, color = TEXT_COLOR, other_angle = False, quadrant = (1, -1))
        self.add(angle_c)
        angle_c_annotation = MathTex(r"\theta_c", color = TEXT_COLOR).scale(0.7)
        angle_c_annotation.next_to(angle_c, buff = 0.2)
        self.add(angle_c_annotation)

        ray_2_start = ray_1_end + 0.1*UP + 0.1*LEFT
        ray_2_end = 2*RIGHT + 0.2*DOWN
        ray_2 = Arrow(start = ray_2_start, end = ray_2_end, color = ARROW_COLOR)
        self.add(ray_2)
        angle_f = Angle(self.zero_line, ray_2, radius = 0.5, color = TEXT_COLOR, other_angle = False, quadrant = (1, -1))
        self.add(angle_f)
        angle_f_annotation = MathTex(r"\theta_f", color = TEXT_COLOR).scale(0.7)
        angle_f_annotation.next_to(angle_f, buff = 0.2)
        self.add(angle_f_annotation)

        angle = Angle(ray_1, dashed_line_1, radius = 1, color = TEXT_COLOR, other_angle = False, quadrant = (-1, 1))
        self.add(angle)
        angle_annotation = MathTex(r"\theta", color = TEXT_COLOR).scale(0.7)
        angle_annotation.next_to(angle, buff = 0.2)
        self.add(angle_annotation)

        R_arrow = DoubleArrow(start = ORIGIN + 0.1*LEFT, end = 5*RIGHT, color = BLACK).shift(0.2*DOWN)
        R_arrow_annotation = MathTex(r"R", color = TEXT_COLOR).scale(0.7)
        R_arrow_annotation.next_to(R_arrow.get_center(), buff = 0.2).shift(0.2*DOWN)
        self.add(R_arrow)
        self.add(R_arrow_annotation)

        y_line = DoubleArrow(start = ray_1.get_end()+0.3*UP, end = 4*RIGHT + 0.3*DOWN, color = BLACK).shift(0.5*RIGHT)
        y_line_annotation = MathTex(r"y_i=y_f", color = TEXT_COLOR).scale(0.7)
        y_line_annotation.next_to(y_line.get_center(), buff = 0.2).shift(0.2*DOWN)
        self.add(y_line)
        self.add(y_line_annotation)
        # self.add(y_line)

```
